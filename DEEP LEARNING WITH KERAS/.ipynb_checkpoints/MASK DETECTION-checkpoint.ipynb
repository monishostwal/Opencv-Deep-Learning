```python
import keras
import cv2
import matplotlib.pyplot as plt
```

```python
masked=cv2.imread('./Datasets/MASK_DETECTION/train/with_mask/0-with-mask.jpg')
masked=cv2.cvtColor(masked,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(masked)
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
img_generator=ImageDataGenerator(rotation_range=30,width_shift_range=0.2,height_shift_range=0.2,fill_mode='nearest',zoom_range=0.2,horizontal_flip=True,shear_range=0.2,rescale=1/255)
```

```python
train_data_gen=img_generator.flow_from_directory('./Datasets/MASK_DETECTION/train/',target_size=(224,224),class_mode='binary',batch_size=16)
```

```python
test_data_gen=img_generator.flow_from_directory('./Datasets/MASK_DETECTION/test/',target_size=(224,224),class_mode='binary',batch_size=16)
```

```python
train_data_gen.class_indices
```

```python
from keras.models import Sequential
```

```python
model=Sequential([
                    keras.layers.Conv2D(32,(3,3),activation='relu',input_shape=(224,224,3)),
                    keras.layers.MaxPooling2D(pool_size=(2,2)),
                    keras.layers.Conv2D(64,(3,3),activation='relu'),
                    keras.layers.MaxPooling2D(pool_size=(2,2)),
                    keras.layers.Flatten(),
                    keras.layers.Dense(64,activation='relu'),
                    keras.layers.Dense(128,activation='relu'),
                    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit_generator(train_data_gen,epochs=10,validation_data=test_data_gen)
```

```python
model.evaluate_generator(test_data_gen)
```

```python
model.save('mask_model.h5')
```

# Reading video from opencv

```python
from keras.models import load_model
def model_predict(img):
    model=load_model('./mask_model.h5')
    class_number=model.predict_classes(img)
    return class_number[0][0]
    
```

```python
def detect_mask(img):
    gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    face_classifier=cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
    faces=face_classifier.detectMultiScale(gray_img,1.1,3)
    faceimg=img
    if faces is ():
        print("No face Found")
        return img
    else:
        for ( x,y,w,h) in faces:
            r = max(w, h) / 2
            centerx = x + w / 2
            centery = y + h / 2
            nx = int(centerx - r)
            ny = int(centery - r)
            nr = int(r * 2)
            faceimg = img[ny:ny+nr, nx:nx+nr]
            faceimg=cv2.resize(faceimg,(224,224))
            faceimg=faceimg.reshape((1,224,224,3))
            faceimg=faceimg/255.0
            class_number=model_predict(faceimg)
            print(class_number)
            string=''
#             if(class_number==0):
#                 string='Mask is present'
#                 new_img=cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
#             else:
#                 string='Mask is not present'
#                 new_img=cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),3)
            new_img=cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)
                
    return new_img
```

```python
import cv2
cap=cv2.VideoCapture(0)
while(True):
    ret,frame=cap.read()
    
    cv2.imshow("Video",detect_mask(frame))
    if(cv2.waitKey(1)==27):
        break
cv2.destroyAllWindows()
cap.release()
```
